## Copying vs Viewing 
In instances whereby we are working with large amounts of data, it may be slower and more memory heavy to evoke a standard function which "copies" the data over, rather than embedding a @view macros inside the code and "lazily viewing" it - this sounds confusing so an example is needed

```julia
# using the view function 
[mean(@view $x[:, j]) for j in axes($x, 2)] 
# compared to an ordinary function 
[mean($x[:, j]) for j in axes($x, 2)] 
```

Using BenchMark tools, the first @view function took 40ms and used 144bytes of memory, whereas the second function too 200ms and used 762mB of memory - an astonishing difference indeed!!! - These are simple introductory examples which don't have much context, but keep this in mind if you ever approach problems whereby speed and memory are critically important; also, it's just good code hygiene/best practice

The optimized julia function `mean(x, dims=1)` is already super quick and efficient, and so should be used by default

## Fitting a linear regression
We can now turn to fitting a linear regression by using ordinary least squares (OLS). Later you will learn a more general and convenient API for estimating parameters of such models using the GLM.jl package, but for now, we restrict ourselves to a basic approach to learn how to work with matrices in Julia. The linear regression that we want to fit has the form y = a + b * x + error, where a and b are unknown coefficients that must be estimated. We will choose them so that the sum of squares of error terms across all observations is minimized.

Nice overview of a linear regression https://www.youtube.com/watch?v=7ArmBVF2dCs 

Generate a vector of values for our response/target variable, and a vector for our explanatory variable x. 

````julia
y = aq[:, 2]
````

````julia
X = [ones(11) aq[:,1]]
````

"Then we use the method of constructing matrices in which you merge columns by separating them with a whitespace and wrapping them in square brackets. This is the same approach we used at the beginning of this section. The only difference is that now we are horizontally concatenating whole vectors, not just single cells of a matrix."

**"Fitting a linear model to a given data set usually requires estimating the regression coefficients β such that the error term ε = y − X β is minimized."** From wikipedia 

````julia
# Estimate β (regression co-efficients) and ϵ (error) 
X \ y 
````

The **\\** operator is used to estimate the parameters of the model

Now estimate this for all the models

````julia
[[ones(11) aq[:, i]] \ aq[:, i+1] for i in 1:2:7]
````

They are all virtually the same models

Now compute the R² coefficient of determination.

````julia
function R²(x, y)
          X = [ones(11) x]
          model = X \ y 
          prediction = X * model 
          error = y - prediction 
          SS_res = sum(v -> v ^ 2, error)
          mean_y = mean(y)
          SS_tot = sum(v -> (v - mean_y) ^ 2, y)
          return 1 - SS_res / SS_tot
      end
````

````julia
[R²(aq[:, i], aq[:, i+1]) for i in 1:2:7]
````

They are all virtually the same!

## Mapping key-value pairs with dictionaries
Another type of standard collection often used when doing data science is the dictionary. We will introduce dictionaries by solving the famous Sicherman dice puzzle. In this section, you will learn how to create a dictionary, add keys to a dictionary, retrieve values from it, and compare dictionaries.


## The Sicherman puzzle

A standard die has six sides, with faces numbered from 1 to 6. In many games, players roll two standard dice and sum the obtained results. In the Sicherman puzzle, we are asked to check if it is possible to number the faces of a pair of cubes in a way completely different from that of standard dice so that the cubes can be used in any dice game and all the odds will be the same as they are when standard dice are used. More formally, we want to check if there exist other pairs of two six-sided dice, not necessarily identical, with faces numbered with positive integers that have the same probability distribution for the sum of rolled values as standard dice.

To solve this puzzle, we will use dictionaries. A dictionary is a mapping of keys to values. In our example, since we consider two six-sided dice, we can have 36 distinct toss results (we have six sides of each die, and 6 × 6 = 36 possible combinations of outcomes). Therefore, our mapping will tell us, for every value of the sum of rolled values, how many times (out of the total 36) it occurs.


Let's create a dict of all the possible sums of the two-dice combinations e.g. 1+6, 1+5, 1+4 etc.


````julia
two_standard = Dict{Int, Int}()
````

````julia
for i in [1, 2, 3, 4, 5, 6]
    for j in [1, 2, 3, 4, 5, 6]
        s = i + j
        if haskey(two_standard, s)
            two_standard[s] += 1
        else
            two_standard[s] = 1
        end
    end
end
````

````julia
two_standard
````

Plot the dict values - we have to collect them in order to shape into a vector to be passed to the scatter function

````julia
scatter(collect(keys(two_standard)),
              collect(values(two_standard));
              legend=false, xaxis=2:12)
````

Since julia is magic, the scatter function also allows us to provide the dict unmodified - 
`scatter(two_standard;legend=false, xaxis=2:12)`

## Solving the puzzle

Refer directly to the notes for the entire explanation - in essence what this task involves is creating a vector of all possible 2 die combinations, whereby each side of the dice takes on all possible combinations from 1:11. So for instance, one 2 die combination would be [1, 1, 1, 1, 1] and [1,1,1,1,2] - and from ALL of these possible combinations, we are looking for a two die combination which has the exact same distribution of 2 die sums as our conventional 1:6 dice, effectively allowing us to play any die games using these modified die - same statistical properties.    

````julia
all_combinations = [[1, x2, x3, x4, x5, x6]
    for x2 in 2:11
    for x3 in x2:11
    for x4 in x3:11
    for x5 in x4:11
    for x6 in x5:11] 
````

Now let's run some loops and see if there are any die pairs which satisfy our condition s

````julia
for d1 in all_combinations, d2 in all_combinations
          test = Dict{Int, Int}()
          for i in d1, j in d2
              s = i + j
              if haskey(test, s)
                  test[s] += 1
              else
                  test[s] = 1
              end
          end
          if test == two_standard
              println(d1, " ", d2)
          end
      end
````

## Exercise 4.2
Rewrite the code solving the Sicherman puzzle, wrapping the logic of the processing in functions. Create one function, dice_ distribution, that produces a dictionary with a distribution of the sum of possible combinations of values on two dice passed as its arguments. Next, write another function, test_dice, in which you create the all_dice variable, then the two_standard variable, and finally run the main loop comparing the distribution of all dice from the all_dice vector against the two_ standard distribution.

````julia
function dice_distribution(a, b) 
    dice_dict = Dict()
    for dice1 in a, dice2 in b 
        dice_sum = dice1 + dice2 
        if haskey(dice_dict, dice_sum)
            dice_dict[dice_sum] += 1 
        else 
            dice_dict[dice_sum] = 1 
        end
    end 
    return dice_dict        
end     
````

````julia
dice_distribution([1,2,3,4,5,6], [1,2,3,4,5,6]) 
````
